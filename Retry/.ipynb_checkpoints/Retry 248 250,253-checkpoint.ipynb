```python
test=pd.read_csv('Bank-data-testing.csv')
test_data=test.copy()
test_data
```

```python
test_data['y']=test_data['y'].map({'no':0,'yes':1})
test_data=test_data.drop(['Unnamed: 0'],axis=1)
test_data
```

```python
test_data_accuracy=test_data['y']
test_data=test_data[estimators]
test_data1=sm.add_constant(test_data)
test_data1
```

```python
def confusion_matrix(data,actual_values,model):
        
        #ロジットモデルを使った値の予測
        pred_values = model.predict(data)
        # 階級の規定
        bins=np.array([0,0.5,1])
        # ヒストグラムを作成し、値が0と0.5の間の場合は0に、
        # 0.5と1の間の場合は1にします
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        # 正確性の計算
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        # 混同行列と正確性
        return cm, accuracy
```

```python
# テストデータにおける混同行列の作成
cm = confusion_matrix(test_data1,test_data_accuracy,results_logit)
cm
```

```python
# 訓練データにおける混同行列の作成
cm = confusion_matrix(X,y,results_logit)
cm
```
