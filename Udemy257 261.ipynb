```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('Countries-exercise.csv')
data
```

```python
data_position=data.iloc[:,1:3]
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
```

```python
kmeans=KMeans(3)
```

```python
kmeans.fit(data_position)
```

```python
cluster_identify=kmeans.predict(data_position)
```

```python
data_with_cluster=data.copy()
data_with_cluster['cluster']=cluster_identify
```

```python
plt.scatter(data_with_cluster['Longitude'],data_with_cluster['Latitude'],c=cluster_identify,cmap='rainbow')
plt.savefig('KMeans7')
```

```python
wcss=[]

for i in range(1,11):
    kmeans=KMeans(i)
    kmeans.fit(data_position)
    wcss_iter=kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss_number=range(1,11)
plt.plot(wcss_number,wcss)
```
