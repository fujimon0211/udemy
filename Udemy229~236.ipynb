### 入力とターゲットの宣言

```python
target=data_prosecced['log_price']
input=data_prosecced.drop(['log_price'],axis=1)
```

### データの標準化

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(input)
input_scaled=scaler.transform(input)
```

### 訓練データとテストデータの分割

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_scaled,target,test_size=0.2,random_state=365)
```

### 回帰の作成

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat=reg.predict(x_train)
```

### 散布図で可視化し検証。今回の場合大きなズレはなさそう

```python
plt.scatter(y_hat,y_train)
```

```python
sns.displot(y_hat-y_train)
```

### 重みとバイアス

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(x_train,y_train)
```

```python
reg_summary=pd.DataFrame(index=input.columns)
reg_summary['weights']=reg.coef_
reg_summary
```

### テスト

```python
y_hat_test=reg.predict(x_test)
plt.scatter(y_hat_test,y_test)
```

```python
df_pf=pd.DataFrame(np.exp(y_hat_test),columns=['Prediction'])
```

```python
y_test=y_test.reset_index(drop='True')
df_pf['Target']=np.exp(y_test)
df_pf
```

### 残差を調べる

```python
df_pf['Residual']=df_pf['Target']-df_pf['Prediction']
df_pf
```

### 残差の変化率を調べる

```python
df_pf['Difference%']=np.absolute(df_pf['Residual']/df_pf['Target']*10)
df_pf
```

```python
df_pf.describe()
```
