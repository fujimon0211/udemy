```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
# データを読み込みます
raw_data = pd.read_csv('Categorical.csv')
# データの中身を確認します
raw_data
```

```python
data = raw_data.copy()
```

```python
data_mapped = data.copy()
data_mapped['continent'] = data_mapped['continent'].map({'North America':0,'Europe':1,'Asia':2,'Africa':3,'South America':4, 'Oceania':5,'Seven seas (open ocean)':6, 'Antarctica':7})
data_mapped
```

```python
data_continent=data_mapped.iloc[:,3:4]
kmeans=KMeans(4)
kmeans.fit(data_continent)
cluster_idenfier=kmeans.predict(data_continent)
data_mapped_with_cluster=data.copy()
data_mapped_with_cluster['cluster']=cluster_idenfier
```

```python
plt.scatter(data_mapped_with_cluster['Longitude'],data_mapped_with_cluster['Latitude'],c=cluster_idenfier,cmap='rainbow')
plt.savefig('2 clustering iris.png')
```
